# 옷장패션 데이터 정제 보고서

## 1. 데이터 개요
- 행/열: 1,500 × 8
- 주요 컬럼: order_id, customer_age, category, channel, price, quantity, amount, return_amount

## 2. 진단 결과
- 결측: amount 51건(약 3.4%), price 5건(약 0.3%)
- 이상치(IQR): customer_age(999·0, 5건), quantity(200, 1건), amount(50,000,000, 1건)
- 의심되는 결측 유형: amount는 app 채널에 쏠림 → MAR

## 3. 처리 결정과 근거

| 이슈 | 결정 | 근거 | 한계 |
| --- | --- | --- | --- |
| amount 결측 | 채널별 중앙값 대체 | app 채널에 결측이 집중되어 MAR 가설에 부합 | 채널 외 다른 원인은 검토하지 않음 |
| price 결측 | 카테고리별 중앙값 대체 | 결측 비율이 낮고 카테고리별 가격 특성을 유지 | 표본이 적은 카테고리는 통계가 불안정할 수 있음 |
| customer_age 이상치 | NaN 처리 후 중앙값 대체 | 999와 0은 입력 오류로 판단 | 외부 데이터로 검증하지 못함 |
| quantity 이상치 | NaN 처리 후 중앙값 대체 | 일반 소비자의 구매 수량으로 보기 어려운 값 | 실제 대량 구매였을 가능성 존재 |
| amount 극단값 | 유지 + amount_outlier 플래그 생성 | 실제 고액 거래 가능성을 보존 | 별도 분석 필요 |

## 4. 처리 후 검증
- 결측: 0건(설계상 NaN 유지 컬럼 제외)
- customer_age 범위: 5 ~ 60(정상)
- amount_outlier 플래그: 145건 보존

## 5. 후속 권고
- app 채널의 amount 결측 원인을 추가 분석할 필요가 있다.
- amount_outlier 데이터는 실제 도매 거래 여부를 별도로 분석하는 것이 바람직하다.
