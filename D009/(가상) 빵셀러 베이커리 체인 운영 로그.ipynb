# 빵셀러 데이터 정제·집계 보고서

## 1. 데이터 개요
- 출처: (가상) 빵셀러 베이커리 체인 운영 로그
- 기간: 2025-04-01 ~ 2025-06-30 (3개월)
- 원본 행 수: 2,164행
- 정제 후 행 수: 1,784행

## 2. 발견된 품질 문제
- 결측: revenue 컬럼 60건
- 중복: 321건
- 이상치: revenue 음수(-45,000원) 1건
- 표기 혼재:
  - store_id 공백 존재 (" S01" → "S01")
  - item 표기 혼재 ("케익" → "케이크")
  - date 형식 혼재 (2025-04-01, 2025/04/01, 20250401)

## 3. 처리 결정과 근거
1. 중복 데이터 321건 제거
2. store_id 앞뒤 공백 제거(str.strip())
3. item명을 "케익"에서 "케이크"로 통일
4. date_str을 datetime 형식으로 변환
5. revenue가 음수인 환불 데이터는 별도 보관 후 분석에서 제외
6. revenue 결측치 60건 제거

## 4. 주요 KPI 결과
- 월별 매장별 매출을 집계하여 비교
- 가장 높은 월 매출은 S06 매장의 2025년 4월(9,156,000원)
- 상품별 총매출 1위는 쿠키(26,690,000원)
- 상품별 평균 단가는 크루아상이 가장 높음(약 6,946원)

## 5. 한계와 후속 작업
- 환불 데이터는 별도로 분석하여 환불 원인을 파악할 필요가 있다.
- 결측값 발생 원인을 확인하여 입력 오류를 줄일 필요가 있다.
- 날짜 및 상품명 표기 규칙을 통일하여 데이터 품질을 향상시킬 필요가 있다.
